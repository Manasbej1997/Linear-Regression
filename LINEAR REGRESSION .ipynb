```python
# Required library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Importing data sets
Boston=pd.read_csv("C:\\Users\\Hp\\Documents\\Boston.csv")
Boston.head()
```

```python
Boston.info()
```

```python
# Check shape
Boston.shape
```

```python
# show heatmap corr
plt.figure(figsize=(12,10))
sns.heatmap(Boston.corr(),annot=True)
```

```python
# Spliting data set into X and y variable
X=Boston[['crim']]
y= Boston[['medv']]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Spliting  data sets into trianing and testing data sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
lr= LinearRegression()
```

```python
# fitting the model into X_train and y_train
lr.fit(X_train,y_train)
```

```python
# Predict the values
y_pred=lr.predict(X_test)
```

```python
y_test.head()
```

```python
y_pred[0:5]
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('crim')
plt.ylabel('medv')
plt.title('Boston_scatter_plot')
```
